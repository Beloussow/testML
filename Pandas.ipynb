```python
import sqlite3
import pandas as pd

con = sqlite3.connect('K_F_M.db')

kfc_data = pd.read_sql_query('SELECT * FROM KFC', con)
Burgerking_data = pd.read_sql_query('SELECT * FROM Burgerking', con)
Mcdonalds_data = pd.read_sql_query('SELECT * FROM Mcdonalds', con)
```

```python
# kfc_data = kfc_data.sort_values(by=['city']) #Сортировка
kfc_data_M = kfc_data[kfc_data['city']==' Москва']
kfc_data_M
```

```python
# Burgerking_data = Burgerking_data.sort_values(by=['city']) #Сортировка
Burgerking_data_M = Burgerking_data[Burgerking_data['city']=='Москва']
Burgerking_data_M
```

```python
# Mcdonalds_data = Mcdonalds_data.sort_values(by=['city']) #Сортировка
Mcdonalds_data_M = Mcdonalds_data[Mcdonalds_data['city']=='Москва']
Mcdonalds_data_M
```

```python
# Москва

number_of_restaurants_m = len(kfc_data_M)+len(Burgerking_data_M)+len(Mcdonalds_data_M)

print('KFC', len(kfc_data_M), '-', len(kfc_data_M)/number_of_restaurants_m*100)
print('Burgerking', len(Burgerking_data_M), '-', len(Burgerking_data_M)/number_of_restaurants_m*100)
print('Mcdonalds', len(Mcdonalds_data_M), '-', len(Mcdonalds_data_M)/number_of_restaurants_m*100)
print('Всего', number_of_restaurants_m, '-', '100%')
```

```python
# Всего

number_of_restaurants = len(kfc_data)+len(Burgerking_data)+len(Mcdonalds_data)

print('KFC', len(kfc_data), '-', len(kfc_data)/number_of_restaurants*100)
print('Burgerking', len(Burgerking_data), '-', len(Burgerking_data)/number_of_restaurants*100)
print('Mcdonalds', len(Mcdonalds_data), '-', len(Mcdonalds_data)/number_of_restaurants*100)
print('Всего', number_of_restaurants, '-', '100%')
```
